# Relation entre Décès Maternels et Audits des Décès Maternels

# Définition des Décès Maternels :
Les décès maternels sont définis comme la mort d'une femme pendant la grossesse ou dans les 42 jours suivant la fin de la grossesse, indépendamment de la durée et du site de la grossesse, due à des causes liées à la grossesse ou à sa gestion, mais non à des causes accidentelles ou fortuites.

# Définition des Audits des Décès Maternels :
Les audits des décès maternels sont un processus systématique d'examen des circonstances entourant chaque décès maternel. Ils impliquent une analyse détaillée des dossiers médicaux et des entretiens avec les personnes impliquées dans les soins de la patiente. L'objectif est de comprendre les facteurs contributifs, d'identifier les lacunes dans les soins, et de formuler des recommandations pour prévenir de futurs décès similaires.

# Relation entre les Deux Concepts
# Objectif des Audits :

Les audits des décès maternels visent à identifier les causes et les facteurs contributifs aux décès maternels. En examinant chaque décès en détail, les audits permettent de comprendre pourquoi et comment un décès maternel s'est produit.
# Amélioration de la Qualité des Soins :

Les résultats des audits fournissent des informations précieuses sur les points faibles des systèmes de santé. En identifiant les erreurs et les retards dans les soins, les audits aident à mettre en place des mesures correctives pour améliorer la qualité des soins maternels.
# Éducation et Sensibilisation :

Les audits des décès maternels sensibilisent les professionnels de santé aux erreurs courantes et aux meilleures pratiques. Ils offrent des opportunités d'apprentissage pour le personnel médical, favorisant ainsi une culture d'amélioration continue.
# Politiques et Interventions :

Les données issues des audits des décès maternels servent de base pour l'élaboration de politiques de santé publique. Les recommandations des audits peuvent mener à des changements dans les protocoles de soins, la formation des personnels et les allocations de ressources.
Réduction de la Mortalité Maternelle :

En identifiant les causes et les circonstances des décès maternels, les audits permettent de mettre en place des interventions ciblées pour prévenir de futurs décès. Ils jouent un rôle crucial dans la réduction de la mortalité maternelle en fournissant des preuves concrètes sur les domaines nécessitant des améliorations.
# Suivi et Évaluation :

Les audits des décès maternels permettent de suivre les progrès réalisés dans la réduction des décès maternels. En répétant régulièrement ces audits, les systèmes de santé peuvent évaluer l'efficacité des interventions mises en place et ajuster les stratégies en conséquence.
# Conclusion
Les audits des décès maternels sont une méthode essentielle pour comprendre les raisons des décès maternels et pour améliorer la qualité des soins de santé maternelle. Ils fournissent des informations précieuses qui peuvent être utilisées pour prévenir de futurs décès, éduquer les professionnels de santé, influencer les politiques de santé et, finalement, réduire la mortalité maternelle. La relation entre les décès maternels et les audits des décès maternels est donc intrinsèque, l'un étant un moyen de mieux comprendre et de réduire l'autre.

```python
# The boxplot visually represents the distribution of the number of audits. Here's how to interpret it:

# - The box itself represents the interquartile range (IQR), which contains the middle 50% of the data. The bottom of the box is the 25th percentile (Q1), and the top of the box is the 75th percentile (Q3).
# - The line inside the box represents the median (50th percentile).
# - The "whiskers" extending from the box represent the range of the data, excluding outliers. They typically extend to 1.5 times the IQR from the edges of the box.
# - Individual points outside the whiskers are considered outliers.

# In this specific plot, we can observe:

# - The median number of audits is around the middle of the box.
# - The IQR (the height of the box) indicates the spread of the middle 50% of the data.
# - The presence of outliers above the upper whisker suggests that there are a few municipalities with a significantly higher number of audits than the rest.

# Overall, the boxplot provides a quick visual summary of the distribution of the number of audits, highlighting the central tendency, spread, and the presence of any outliers.
```

Median Around the Middle of the Box:

The line inside the box represents the median, which is the middle value when the data is ordered from least to greatest.
In your case, the median being around the middle of the box suggests that the distribution of the number of audits is roughly symmetrical. This means there are roughly the same number of municipalities with audit counts above and below the median.
2. IQR (Interquartile Range):

The height of the box represents the IQR, which is the range between the 25th percentile (bottom of the box) and the 75th percentile (top of the box).
The IQR tells us about the spread of the middle 50% of the data. A larger IQR indicates more variability in the middle of the distribution.
In your case, the IQR gives you an idea of the typical range of audit counts for most municipalities.
3. Outliers Above the Upper Whisker:

The "whiskers" of the box plot extend to a certain distance from the box, usually calculated as 1.5 times the IQR.
Points beyond the whiskers are considered potential outliers, meaning they are significantly higher or lower than the majority of the data.
In your case, the outliers above the upper whisker indicate that there are a few municipalities with a much higher number of audits compared to the rest. This suggests that those municipalities might have unique characteristics or circumstances that lead to more frequent audits.

```python
# prompt: remove missing values

# Drop rows with any missing values
df_cleaned = df.dropna()

# Display the cleaned DataFrame
print(df_cleaned)
```

```python
# Convert 'Number of Maternal Death' to numeric, coercing errors
df_cleaned['Number of Maternal Death'] = pd.to_numeric(df_cleaned['Number of Maternal Death'], errors='coerce')

# Calculate the IQR for 'Number of Maternal Death'
Q1 = df_cleaned['Number of Maternal Death'].quantile(0.25)
Q3 = df_cleaned['Number of Maternal Death'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identify outliers
outliers = df_cleaned[(df_cleaned['Number of Maternal Death'] < lower_bound) | (df_cleaned['Number of Maternal Death'] > upper_bound)]

print(outliers)
```

The results you've shown indicate the rows in your DataFrame (df_cleaned) that were identified as outliers based on the 'Number of Maternal Death' column.

Here's a breakdown of what the output means:

Year: The year in which the data was recorded.
Municipalities: The specific municipality within Burundi where the data was collected.
Number of Maternal Death: The number of maternal deaths recorded in that municipality for that year.
Why are these rows considered outliers?

Outliers are data points that significantly deviate from the rest of the data. In this case, the outliers were identified using the Interquartile Range (IQR) method. This method calculates the range between the 25th percentile (Q1) and the 75th percentile (Q3) of the data. Any data point that falls below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR is considered an outlier.

What does this tell us?

The presence of these outliers suggests that in certain years and municipalities, there were unusually high numbers of maternal deaths compared to the general trend in the data. This could indicate:

Specific events: Potential factors like disease outbreaks, natural disasters, or conflicts might have occurred in those years or municipalities, leading to increased maternal mortality.
